```python
df_train['Fault_Code'] = df_train[target_cols].astype(str).sum(axis=1)
df_train.groupby(['Fault_Code']).size()
```

```python
# Crate Fault Category column
df_train['Fault_Category'] = 'NA'
df_train.loc[df_train['Fault_Code'] == '0000000', 'Fault_Category'] = 'Pass'
df_train.loc[df_train['Fault_Code'] == '0000001', 'Fault_Category'] = 'Other Faults'
df_train.loc[df_train['Fault_Code'] == '0000010', 'Fault_Category'] = 'Bumps'
df_train.loc[df_train['Fault_Code'] == '0000100', 'Fault_Category'] = 'Dirtiness'
df_train.loc[df_train['Fault_Code'] == '0001000', 'Fault_Category'] = 'Stains'
df_train.loc[df_train['Fault_Code'] == '0010000', 'Fault_Category'] = 'K_Scratch'
df_train.loc[df_train['Fault_Code'] == '0100000', 'Fault_Category'] = 'Z_Scratch'
df_train.loc[df_train['Fault_Code'] == '1000000', 'Fault_Category'] = 'Pastry'
df_train.loc[df_train['Fault_Code'] == '0010001', 'Fault_Category'] = 'K_Scratch,Other_Faults'
df_train.loc[df_train['Fault_Code'] == '0010010', 'Fault_Category'] = 'K_Scratch,Bumps'
df_train.loc[df_train['Fault_Code'] == '1010000', 'Fault_Category'] = 'K_Scratch,Pastry'
df_train_faultsum = df_train.groupby(['Fault_Code','Fault_Category']).size().reset_index(name='count')
df_train_faultsum
```

```python
# Pareto of Faults
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.ticker import PercentFormatter

df_train_faultsum = df_train_faultsum.sort_values(by='count',ascending=False).reset_index(drop=True)
df_train_faultsum["cumulative_pct"] = df_train_faultsum['count'].cumsum()/df_train_faultsum['count'].sum()*100
df_train_faultsum

```

```python
# Plot Pareto
fig, ax = plt.subplots()
ax.bar(df_train_faultsum.index, df_train_faultsum['count'], color="C0")
ax2 = ax.twinx()
ax2.plot(df_train_faultsum['Fault_Category'], df_train_faultsum["cumulative_pct"], color="C1", marker="D", ms=7)
ax2.yaxis.set_major_formatter(PercentFormatter())
ax.tick_params(axis="y", colors="C0")
ax2.tick_params(axis="y", colors="C1")
for tick in ax.get_xticklabels():
        tick.set_rotation(45)
plt.show()
```
